```python
import pandas as pd
# Definir os dados em listas
anos = [1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
salario_minimo = [120, 130, 136, 151, 180, 200, 240, 260, 300, 350, 380, 415, 465, 510, 545, 622, 678, 724, 788, 880, 937, 954, 998, 1039.00, 1100.00, 1212.00, 1320.00]
# Remover a vírgula e converter para tipo float para os salários de 2020 em diante
salario_minimo = [float(str(salario).replace(',', '').replace('.', '')) if isinstance(salario, str) else salario for salario in salario_minimo]
# Criar um DataFrame
df = pd.DataFrame({'Salario_minimo': salario_minimo}, index=anos)
# Imprimir o DataFrame
print(df[-6:])
```

```python
# Transpor (pivotar) o DataFrame
df_transposto = df.T

# Imprimir o DataFrame transposto
df_transposto
```
