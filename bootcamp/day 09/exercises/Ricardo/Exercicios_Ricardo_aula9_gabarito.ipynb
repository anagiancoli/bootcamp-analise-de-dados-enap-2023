# Faça a combinação do dataframe "df" (com informações das trabalhadoras) com o dataframe "estados", de forma que para cada observação referente a uma trabalhadora seja acrescentada a informação de PIB per capta do estado que ela trabalha. Utilize o método how='inner'.

A chave para combinação é a variável uf.

```python
pd.merge(df, estados, how='inner')
# Inner: Utiliza somente as combinacões de chaves observadas nas duas tabelas sendo combinadas
```

# Nesse caso, faria diferença usar o Left Outer Join ou o Right Outer Join ou o Outer Join?

```python
# Nesse caso, como temos todas as chaves em ambas as tabelas, o resultado seria o mesmo com todos os métodos
```

```python

pd.merge(df, estados, how='left')
#Left Outer Join utiliza todas as combinacões de chaves encontradas na tabela à esquerda
```

```python
pd.merge(df, estados, how='right')
#Right Outer Join utiliza todas as combinacões de chaves encontradas na tabela à direita
```

```python
pd.merge(df, estados, how='outer')
#Outer Join utiliza todas as combinacões de chaves observadas nas duas tabelas em conjunto
```

# Faça a combinação do dataframe amostra com o dataframe estados, de forma que para cada observação referente a uma trabalhadora seja acrescentada a informação de PIB per capta do estado que ela trabalha.


Qual tipo de merge você deve usar dessa vez?


Observe que dessa vez temos mais informações de estados do que de trabalhadoras (alguns estados não são residencia de nenhuma trabalhadora da amostra). Se utilizarmos o outer join ou o right join, nossa base de dados será 'poluída' com informações de UF e PIBPC_estado, mesmo sem existirem trabalhadoras na base de dados associadas a eles.

Utilizando o outer join ou o right join, a nova base de dados após o merge, nos passará a falsa impressão de que teremos mais trabalhadoras do que realmente temos na nossa base de dados.

```python
pd.merge(amostra, estados, how='left')
```

### O que acontece se você utilizar o Outer Join?

```python
pd.merge(amostra, estados, how='outer')
```
